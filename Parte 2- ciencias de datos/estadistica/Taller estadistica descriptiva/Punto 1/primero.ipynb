```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datasetDD = pd.read_excel("Datos_01.xlsx")
sheet_name="Edades"
```

```python
datasetDD.info()
```

```python
frecuenciaCD = datasetDD["Edades"].value_counts().reset_index()
frecuenciaCD.info()
```

```python
frecuenciaCD.columns=["Edades", "Frecuecia_absoluta"]
frecuenciaCD
```

```python
frecuenciaCD["Frecuecia_acumulada_absoluta"] = frecuenciaCD["Frecuecia_absoluta"].cumsum()

frecuenciaCD
```

```python
suma_absoluta = frecuenciaCD["Frecuecia_absoluta"].sum()
frecuenciaCD["Frecuencia_relativa"] = frecuenciaCD["Frecuecia_absoluta"] / suma_absoluta

frecuenciaCD
```

```python
frecuenciaCD["Frecuecia_acumulada_realtiva"] = frecuenciaCD ["Frecuencia_relativa"].cumsum()
frecuenciaCD
```

```python
media = frecuenciaCD['Edades'].mean()
mediana = frecuenciaCD['Edades'].median()
minimo = frecuenciaCD['Edades'].min()           #Cuantitativas
maximo = frecuenciaCD['Edades'].max()
moda = frecuenciaCD['Edades'].mode().tolist()   #Cualitativa

print(f"Media: {media}")
print(f"Mediana: {mediana}")
print(f"Mínimo: {minimo}")
print(f"Máximo: {maximo}")
print(f"Moda: {moda}")
```

Graficos


```python
plt.bar(frecuenciaCD['Edades'], frecuenciaCD['Frecuecia_absoluta'], color='skyblue')
plt.xlabel('Edades')
plt.ylabel('Frecuencia Absoluta')
plt.title('Frecuencia Absoluta por Edades')
plt.show()
```

```python
total_frecuencia = frecuenciaCD['Frecuecia_absoluta'].sum()
frecuenciaCD['Frecuencia_relativa'] = frecuenciaCD['Frecuecia_absoluta'] / total_frecuencia

# Crear el gráfico de pastel
plt.figure(figsize=(8, 8))  # Tamaño del gráfico
plt.pie(frecuenciaCD['Frecuencia_relativa'], labels=frecuenciaCD['Edades'], autopct='%1.1f%%', colors=plt.get_cmap('Pastel1').colors)

plt.title('Distribución de Frecuencia Relativa por Estaturas')
plt.show()
```

```python
frecuenciaCD['Frecuencia_acumulada'] = frecuenciaCD['Frecuecia_absoluta'].cumsum()

plt.bar(frecuenciaCD['Edades'], frecuenciaCD['Frecuencia_acumulada'], color='lightcoral')
plt.xlabel('Edades')
plt.ylabel('Frecuencia Acumulada Absoluta')
plt.title('Frecuencia Acumulada Absoluta por Edades')
plt.show()
```

```python
frecuenciaCD['Frecuencia_acumulada_absoluta'] = frecuenciaCD['Frecuecia_absoluta'].cumsum()
frecuenciaCD['Frecuencia_acumulada_relativa'] = frecuenciaCD['Frecuencia_acumulada_absoluta'] / frecuenciaCD['Frecuecia_absoluta'].sum()

plt.bar(frecuenciaCD['Edades'], frecuenciaCD['Frecuencia_acumulada_relativa'], color='lightcoral')
plt.xlabel('Edades')
plt.ylabel('Frecuencia Acumulada Relativa')
plt.title('Frecuencia Acumulada Relativa por Edades')

plt.show()  
```

Conclusion

Las edades más comunes en el grupo son 16 y 18 años, cada una con 11 observaciones. 
La frecuencia acumulada indica que hasta los 22 años se concentra el 94% de las observaciones. 
La frecuencia relativa muestra que el 44% de los datos están en edades hasta 18 años, y el 100% hasta 21 años.
