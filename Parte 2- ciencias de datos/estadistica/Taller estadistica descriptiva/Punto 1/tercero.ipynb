```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Leer los datos del archivo Excel
datasetDD = pd.read_excel("Datos_03.xlsx", sheet_name="Hoja1")

# Mostrar el DataFrame original
print("DataFrame Original:")
print(datasetDD)

# Definir un diccionario de mapeo para convertir colores a números
mapeo_colores = {
    'azul': 1,
    'verde': 2,
    'morado': 3,
    'amarillo': 4,
    'rojo': 5,
    'naranja': 6
}

# Crear una nueva columna con los datos convertidos a números
datasetDD['Colores_numericos'] = datasetDD['Colores'].map(mapeo_colores)


```

```python
datasetDD
```

```python
frecuenciaCD = datasetDD["Colores_numericos"].value_counts().reset_index()
frecuenciaCD.info()
```

```python
frecuenciaCD
```

```python
frecuenciaCD.columns=["Colores_numericos", "Frecuecia_absoluta"]
frecuenciaCD
```

```python
frecuenciaCD["Frecuecia_acumulada_absoluta"] = frecuenciaCD["Frecuecia_absoluta"].cumsum()

frecuenciaCD
```

```python
suma_absoluta = frecuenciaCD["Frecuecia_absoluta"].sum()
frecuenciaCD["Frecuencia_relativa"] = frecuenciaCD["Frecuecia_absoluta"] / suma_absoluta

frecuenciaCD
```

```python
frecuenciaCD["Frecuecia_acumulada_realtiva"] = frecuenciaCD ["Frecuencia_relativa"].cumsum()
frecuenciaCD
```

```python
media = frecuenciaCD['Frecuecia_absoluta'].mean()
mediana = frecuenciaCD['Frecuecia_absoluta'].median()
minimo = frecuenciaCD['Frecuecia_absoluta'].min()           #Cuantitativas
maximo = frecuenciaCD['Frecuecia_absoluta'].max()
moda = frecuenciaCD['Frecuecia_absoluta'].mode().tolist()

moda_edades = frecuenciaCD['Colores_numericos'].mode().tolist()        #Cualitativa

print(f"Moda de Colores: {moda_edades}")
print(f"Media: {media}")
print(f"Mediana: {mediana}")
print(f"Mínimo: {minimo}")
print(f"Máximo: {maximo}")
print(f"Moda: {moda}")

# azul: 1
# verde: 2
# morado: 3
# amarillo: 4
# rojo: 5
# naranja: 6
```

Graficos


```python
plt.bar(frecuenciaCD['Colores_numericos'], frecuenciaCD['Frecuecia_absoluta'], color='skyblue')
plt.xlabel('Colores')
plt.ylabel('Frecuencia Absoluta')
plt.title('Frecuencia Absoluta por Colores')
plt.show()

# azul: 1
# verde: 2
# morado: 3
# amarillo: 4
# rojo: 5
# naranja: 6
```

```python
total_frecuencia = frecuenciaCD['Frecuecia_absoluta'].sum()
frecuenciaCD['Frecuencia_relativa'] = frecuenciaCD['Frecuecia_absoluta'] / total_frecuencia
plt.bar(frecuenciaCD['Colores_numericos'], frecuenciaCD['Frecuencia_relativa'], color='lightgreen')
plt.xlabel('Colores')
plt.ylabel('Frecuencia Relativa')
plt.title('Frecuencia Relativa por Colores')
plt.show()

# azul: 1
# verde: 2
# morado: 3
# amarillo: 4
# rojo: 5
# naranja: 6
```

```python
frecuenciaCD['Frecuencia_acumulada'] = frecuenciaCD['Frecuecia_absoluta'].cumsum()

plt.bar(frecuenciaCD['Colores_numericos'], frecuenciaCD['Frecuencia_acumulada'], color='lightcoral')
plt.xlabel('Colores')
plt.ylabel('Frecuencia Acumulada Absoluta')
plt.title('Frecuencia Acumulada Absoluta por Colores')
plt.show()

# azul: 1
# verde: 2
# morado: 3
# amarillo: 4
# rojo: 5
# naranja: 6
```

```python
frecuenciaCD['Frecuencia_acumulada_absoluta'] = frecuenciaCD['Frecuecia_absoluta'].cumsum()
frecuenciaCD['Frecuencia_acumulada_relativa'] = frecuenciaCD['Frecuencia_acumulada_absoluta'] / frecuenciaCD['Frecuecia_absoluta'].sum()

plt.bar(frecuenciaCD['Colores_numericos'], frecuenciaCD['Frecuencia_acumulada_relativa'], color='lightcoral')
plt.xlabel('Colores')
plt.ylabel('Frecuencia Acumulada Relativa')
plt.title('Frecuencia Acumulada Relativa por Colores')

plt.show()

# azul: 1
# verde: 2 
# morado: 3
# amarillo: 4
# rojo: 5
# naranja: 6
```

Conclusion

La frecuencia de los colores numéricos revela que el color representado por el Color azul y el amarillo son los más comunes, cada uno con una frecuencia absoluta de 13. 
Estos colores representan el 23.64% de los datos, acumulando el 47.27% de la frecuencia relativa. 
Los colores verder y morado también tienen una presencia significativa, con frecuencias absolutas de 9 y 8, respectivamente. 
Por otro lado, los colores rojo y naranja tienen una frecuencia absoluta menor, con 6 cada uno. En conjunto, los datos están bien distribuidos, con la frecuencia acumulada alcanzando el 100% al incluir todos los colores.
