```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datasetDD = pd.read_excel("Datosestadisticos.xlsx")
sheet_name="Cuantitativos_Discretos"
```

```python
datasetDD.info()
```

```python
frecuenciaCD = datasetDD["Talla_Zapatos"].value_counts().reset_index()
frecuenciaCD.info()
```

```python
frecuenciaCD.columns=["Talla_zapatos", "Frecuecia_absoluta"]
frecuenciaCD
```

```python
frecuenciaCD["Frecuecia_acumulada_absoluta"] = frecuenciaCD["Frecuecia_absoluta"].cumsum()

frecuenciaCD
```

```python
suma_absoluta = frecuenciaCD["Frecuecia_absoluta"].sum()
frecuenciaCD["Frecuencia_relativa"] = frecuenciaCD["Frecuecia_absoluta"] / suma_absoluta

frecuenciaCD
```

```python
frecuenciaCD["Frecuecia_acumulada_realtiva"] = frecuenciaCD ["Frecuencia_relativa"].cumsum()
frecuenciaCD
```
