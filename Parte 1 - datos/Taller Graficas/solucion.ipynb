```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Diagrama Barras

```python
df = pd.read_csv('dataset_banco2.csv')

plt.figure(figsize=(20, 6))
plt.bar(df['job'], df['balance'])
plt.title('Trabajos y sus salarios')
plt.xlabel('Trabajo')
plt.ylabel('Salario')
plt.show()
```

```python
df = pd.read_csv('dataset_banco2.csv')

plt.figure(figsize=(10, 6))
plt.bar(df['education'], df['balance'])
plt.title('Salario por nivel de educacion')
plt.xlabel('education')
plt.ylabel('Salario')
plt.show()
```

Grafico de Pie

```python
df = pd.read_csv('dataset_banco2.csv')

plt.figure(figsize=(8, 8))
df['marital'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    labels=df['marital'].value_counts().index
)
plt.title('Estado civil')
plt.ylabel('')
plt.show()
```

Grafico de cajon

```python
df = pd.read_csv('dataset_banco2.csv')

data = [df[df['job'] == job]['age'] for job in df['job'].unique()]

plt.figure(figsize=(12, 8))
plt.boxplot(data, tick_labels=df['job'].unique(), vert=True)
plt.title('Trabajos con sus edades')
plt.xlabel('Trabajo')
plt.ylabel('Edad')
plt.xticks(rotation=45)
plt.show()
```

Grafico libre (Dispercion)

```python
df = pd.read_csv('dataset_banco2.csv')

plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['balance'], alpha=0.5)
plt.title('Relación entre edad y salario')
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.show()
```
